```python
b = np.empty([400,400,3])
b.fill(180)
b=b.astype('uint8')

cv2.line(b,(50,100),(100,100),(0,0,0),10)
plt.imshow(b)
```

```python
y = np.zeros((400,400,3),dtype=np.uint8)
y[:,:,0]=0
y[:,:,1]=255
y[:,:,2]=255
plt.imshow(y)
```

```python
font = cv2.FONT_ITALIC
cv2.putText(y,'Loki',(150,200),font,3,(0,0,0),3)
plt.imshow(y)
```
